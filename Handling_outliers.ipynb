## Take Notes

How to handling outliers

IQR: Inter quartile range is difference between 25th and 75th percentile

Box plot lower range is 25 and upper range is 75 while middle is 50th percentile
Upper Outliers: above 75 + 1.5*IQR
Lower Outliers: lower 25 - (1.5*IQR)

```python
salaries = pd.read_csv(r'./datasets/ds_salaries_clean.csv')
salaries.head()

#Calculate percentile of salary
seventy_fifth = salaries['Salary_USD'].quantile(0.75)
print(seventy_fifth)

twenty_fifth = salaries['Salary_USD'].quantile(0.25)
print(twenty_fifth)

IQR = seventy_fifth - twenty_fifth
print(IQR)
upper = seventy_fifth + (1.5 *IQR)
lower = twenty_fifth - (1.5 *IQR)

# Below are outliers records
salaries[(salaries['Salary_USD'] > upper) | (salaries['Salary_USD'] < lower)]\
[["Experience", "Employee_Location", "Salary_USD"]]
```

```python
#dropping outliers

no_outliers = salaries[(salaries['Salary_USD'] > lower) & (salaries['Salary_USD'] < upper)]
no_outliers.describe()
```
